```python
# ==============================================================================
# YOLOv8 Computer Vision Workflow - HIGH RECALL STRATEGY
#
# Description:
# An improved workflow focused on maximizing model recall by implementing
# aggressive augmentation, a more powerful base model, and targeted
# hyperparameter tuning.
# ==============================================================================

import os
import sys
import pandas as pd
from IPython.display import Image, display


# ----------------------------------
# ✅ 1. Configuration - HIGH RECALL SETTINGS
# ----------------------------------
# --- Project and Paths ---
HOME = os.getcwd()
PROJECT_NAME = "yolo_high_recall_attempt" # New project name for this run
DRIVE_MOUNT_PATH = '/content/drive'
DRIVE_PROJECT_PATH = f'{DRIVE_MOUNT_PATH}/MyDrive/yolo' # Root for all YOLO projects
DATASET_YAML_PATH = f'{DRIVE_PROJECT_PATH}/data.yaml'

# --- Model and Training Parameters ---
# ACTION: Switched to a more powerful model (s=small, m=medium, l=large)
MODEL_NAME = 'yolov8s.pt'

# ACTION: Increased epochs for more thorough training
EPOCHS = 200

IMG_SIZE = 227
CONF_THRESHOLD = 0.25 # This is for post-prediction filtering, does not affect training

# --- Prediction Settings ---
PREDICT_SOURCE_DIR = f'{DRIVE_PROJECT_PATH}/data/predict/images'


# (The setup_environment, display_training_results, evaluate_model, and predict_on_images functions remain the same)
def setup_environment():
    """Installs dependencies and mounts Google Drive if in Colab."""
    print("Setting up the environment...")
    os.system('pip install -q "ultralytics>=8.0.130"')
    from ultralytics import YOLO # Moved import inside the function
    import ultralytics # Moved import inside the function
    ultralytics.checks()
    if 'google.colab' in sys.modules:
        print("Detected Google Colab. Mounting Google Drive...")
        from google.colab import drive
        drive.mount(DRIVE_MOUNT_PATH)
    else:
        print("Not running in Google Colab. Skipping Drive mount.")
    print("\nEnvironment setup complete.")
    print("-" * 50)
    # Return the imported classes/modules
    return YOLO, ultralytics


def display_training_results(run_dir):
    """Displays key charts and metrics from the training run."""
    print("Displaying training results...")
    print("\n--- Training Charts ---")
    display(Image(filename=os.path.join(run_dir, 'results.png'), width=800))
    display(Image(filename=os.path.join(run_dir, 'confusion_matrix.png'), width=600))
    display(Image(filename=os.path.join(run_dir, 'val_batch0_pred.jpg'), width=800))
    print("\n--- Final Epoch Metrics ---")
    try:
        results_df = pd.read_csv(os.path.join(run_dir, 'results.csv'))
        results_df.columns = results_df.columns.str.strip()
        print(results_df[['epoch', 'train/box_loss', 'metrics/precision(B)', 'metrics/recall(B)', 'metrics/mAP50(B)']].tail(10))
    except FileNotFoundError:
        print(f"Error: Could not find results.csv in {run_dir}")
    print("-" * 50)

def evaluate_model(weights_path, YOLO_class, ultralytics_module):
    """Evaluates the model on the validation set and prints key metrics."""
    print("Evaluating the final model...")
    model = YOLO_class(weights_path) # Use the passed YOLO class
    metrics = model.val(data=DATASET_YAML_PATH, imgsz=IMG_SIZE) # Use same img size
    print("\n📊 Evaluation Metrics:")
    print(f"   mAP@50-95: {metrics.box.map:.4f}")
    print(f"   mAP@50:    {metrics.box.map50:.4f}")
    print(f"   Precision: {metrics.box.mp:.4f}")
    print(f"   Recall:    {metrics.box.mr:.4f}") # This is the number to watch!
    print("-" * 50)

def predict_on_images(weights_path, YOLO_class):
    """Runs predictions on new images and prints detection details."""
    print(f"Running prediction on images in: {PREDICT_SOURCE_DIR}")
    if not os.path.isdir(PREDICT_SOURCE_DIR):
        print(f"Error: Prediction source directory not found: {PREDICT_SOURCE_DIR}")
        return
    image_files = [f for f in os.listdir(PREDICT_SOURCE_DIR) if os.path.isfile(os.path.join(PREDICT_SOURCE_DIR, f))]
    print(f"Found {len(image_files)} image(s) to predict.")
    model = YOLO_class(weights_path) # Use the passed YOLO class
    results = model.predict(source=PREDICT_SOURCE_DIR, conf=CONF_THRESHOLD, save=True)
    print("\n--- Detection Details ---")
    for i, result in enumerate(results):
        print(f"\nImage {i+1} ({os.path.basename(result.path)}):")
        boxes = result.boxes
        if len(boxes) == 0:
            print("  No objects detected.")
            continue
        for j, box in enumerate(boxes):
            cls_id = int(box.cls[0])
            conf = box.conf[0]
            class_name = result.names[cls_id]
            print(f"  - Detection {j+1}: {class_name} (Confidence: {conf:.2f})")
    if results:
        print(f"\n✅ Prediction complete. Annotated images saved to: {results[0].save_dir}")
    print("-" * 50)
# ----------------------------------
# ✅ 3. (NEW) High-Recall Training Function
# ----------------------------------
def train_for_high_recall(YOLO_class):
    """Trains a YOLOv8 model with a strategy optimized for high recall."""
    print("Starting model training with HIGH-RECALL strategy...")

    # Load the specified base model
    model = YOLO_class(MODEL_NAME) # Use the passed YOLO class

    # Train the model with aggressive augmentation and hyperparameter tuning
    results = model.train(
        data=DATASET_YAML_PATH,
        epochs=EPOCHS,
        imgsz=IMG_SIZE,
        project=DRIVE_PROJECT_PATH,
        name=PROJECT_NAME,
        exist_ok=True,
        plots=True,

        # --- Augmentation Strategy ---
        mosaic=1.0,      # Use mosaic augmentation 100% of the time
        mixup=0.5,       # Use mixup 50% of the time
        copy_paste=0.5,  # Use copy-paste 50% of the time
        fliplr=0.5,      # Horizontal flips
        hsv_h=0.015,     # Image HSV-Hue augmentation (fraction)
        hsv_s=0.7,       # Image HSV-Saturation augmentation (fraction)
        hsv_v=0.4,       # Image HSV-Value augmentation (fraction)

        # --- Hyperparameter Tuning for Recall ---
        # patience=50,   # Default is 50, stops training if no improvement
        # optimizer='AdamW', # AdamW can sometimes perform better
        # cls=1.5,       # Optional: Punish misses more heavily to encourage recall
    )

    latest_run_dir = model.trainer.save_dir
    best_weights_path = os.path.join(latest_run_dir, 'weights/best.pt')

    print(f"\nTraining complete.")
    print(f"📈 Training results saved to: {latest_run_dir}")
    print(f"⚖️ Best model weights saved to: {best_weights_path}")
    print("-" * 50)

    return best_weights_path, latest_run_dir

# ----------------------------------
# ✅ 7. Main Execution Block
# ----------------------------------
if __name__ == '__main__':
    # Get the imported classes/modules from setup_environment
    YOLO_class, ultralytics_module = setup_environment()

    # --- Train the model using the new high-recall strategy ---
    best_weights, training_run_dir = train_for_high_recall(YOLO_class)

    # --- Show training results ---
    display_training_results(training_run_dir)

    # --- Evaluate the final model's performance (Pay close attention to Recall) ---
    evaluate_model(best_weights, YOLO_class, ultralytics_module)

    # --- Run predictions on a folder of new images ---
    predict_on_images(best_weights, YOLO_class)

    print("🎉 Workflow finished successfully! 🎉")
```
